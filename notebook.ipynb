```python
import mach2
```

```python
tree = mach2.Phylogeny.from_file('data/mcpherson_2016/patient1.tree', 'data/mcpherson_2016/patient1.labeling')
```

```python
tree
```

```python
solver = mach2.MACH(tree, primary_site='LOv')
solutions = solver.solve('gurobi', 30, n_threads=30)
```
