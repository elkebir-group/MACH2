```python
import MACH2.src.mach2 as mach2
```

```python
tree = mach2.Phylogeny.from_file('MACH2/data/mcpherson_2016/patient1.tree', 'MACH2/data/mcpherson_2016/patient1.labeling')
```

```python
tree.draw()
```

```python
solver = mach2.MACH(tree, primary_site='LOv')
solutions = solver.solve('gurobi', 30, n_threads=30)
```
