Import MACH2

```python
import mach2
```

Load tree

```python
tree = mach2.MultiLabeledTree.from_files('../data/lung/TRACERxTest.automatic.tree', 
                                 '../data/lung/TRACERxTest.automatic.observed.labeling', 
                                 '../data/lung/TRACERxTest.coloring.txt')
```

Draw tree (use `tree.draw()` to get the graphviz object)

```python
tree
```

Generate solutions

```python
sols = mach2.MACH2(tree, primary_location='p.lung', criteria_ordering='UMCS').solve()
```

Get the number of solutions

```python
len(sols)
```

Examine the refined tree of the first solution

```python
sols[0]
```

Get the parsimony scores of the first solution (for MACH2, every solution in the optimal solution space has the same parsimony score)

```python
sols[0].parsimony_scores()
```

Examine the migration graph of the first solution

```python
sols[0].migration_graph()
```

Examine the summary graph

```python
sols.summary()
```
